```python
import csv
```

```python
import pandas as pd
```

```python
with open('tweets.csv', encoding='utf-8-sig',newline='') as csvfile: 
    reader = csv.DictReader(csvfile)
    count=0
    author=[] #creating a list
    
    for row in reader: #appending the values in the list
        count=count+1
        author.append(row['tweet_author'])
        if count>43348:
            break

acount=[[author.count(item), item] for item in set(author)] #count the occurence of each entity
acount

```

```python
df=pd.DataFrame(acount,columns=['Frequency','Entity'])
```
